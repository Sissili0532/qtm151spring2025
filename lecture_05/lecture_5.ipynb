<font size = "5">

Customized Messages + Numbering

- Initiate "index = 1" before the loop
- Add "index = index + 1" at the end of the body

````python
#------------------------------------------------------------------------------#
# One way to construct a sequence manually, is to gradually add numbers, e.g.
#
#     index = 1
#     index = index + 1
#     index = index + 1
#     index = index + 1
#     etc
#
#     We will do this automatically for a for loop!
#
#     Note: We can use "str(index)" to convert this to a string
#
#------------------------------------------------------------------------------#


list_ids = ["KIA", "Ferrari", "Ford", "Tesla"]

index = 1
print('We are out of the loop', index)


for id in list_ids:
    print("Dear customer, your position is " + str(index) + " on the waitlist" +
           " and your car brand is " + id )
    index = index + 1
    print('We are inside the loop', index)

````

Math Operations + Numbering (Filling)


$ y = x^2 + 2x$

````python
# Create a list of x-values list_x = [1,2,4,5, ..., 50]
# Create a list of y-values to fill in later.
# The "range(

list_x = [1,2,4,5,6,7,8,9,10]
list_y = [None] * len(list_x)

# Create an index 
index = 0
for x in list_x:
    list_y[index] = list_x[index]**2 + 2*list_x[index]
    index = index + 1

# Display results visually
print(list_y)
plt.scatter(list_x, list_y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
````

$ y = x^2 + 2x$

````python
# Create a list of x-values list_x = [1,2,4,5, ..., 50]
# Start an empty list of y-values with []

list_x = [1,2,4,5,6,7,8,9,10]
list_y = []

# Create an index 
for x in list_x:
    y = x**2 + 2*x
    list_y.append(y)

# Display results visually
print(list_y)
plt.scatter(list_x, list_y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
````

````python
# Create a list with the ".csv" dataset names (including their relative subfolder "data_quiz/)
# Run a for-loop over the list of dataset names. 
# Inside the loop: Read the corresponding ".csv" dataset.
# Print the mean of the variable "percent_ages0to14" in each dataset

# Note: This exercise tells you how much the proportion of the population that is between 0 and 14 has varied
# from year to year using real data from the World Development Index.
````

````python
import pandas as pd

# Step 1: Create a list of dataset names
datasets = [
    "data_quiz/wdi_ageprop_2000.csv",
    "data_quiz/wdi_ageprop_2005.csv",
    "data_quiz/wdi_ageprop_2010.csv",
    "data_quiz/wdi_ageprop_2015.csv",
    "data_quiz/wdi_ageprop_2020.csv"
]

# Step 2: Loop over dataset names
for file in datasets:
    # Read CSV file
    df = pd.read_csv(file)
    
    # Compute mean of "percent_ages0to14"
    mean_value = df["percent_ages0to14"].mean()
    
    # Print the result with dataset year
    print(f"Mean of percent_ages0to14 in {file}: {mean_value:.2f}")
````
